```python
# In notebook 02_rule_engine.ipynb:

import pandas as pd
from src.rule_engine import RuleEngine

df = pd.read_csv('../data/referral_dataset.csv')
engine = RuleEngine()

results = df.apply(lambda row: engine.evaluate(row.to_dict()), axis=1)
df['rule_score'] = results.apply(lambda x: x['rule_score'])
df['rules_triggered'] = results.apply(lambda x: str(x['rules_triggered']))
df['is_hard_block'] = results.apply(lambda x: x['is_hard_block'])

# Evaluation
from sklearn.metrics import classification_report
rule_pred = (df['rule_score'] > 0.5).astype(int)
print(classification_report(df['is_fraud'], rule_pred))

# Check rule coverage
print('Fraud caught by rules:', df[df['is_fraud']==1]['rule_score'].gt(0.5).mean())
print('Legit wrongly flagged:', df[df['is_fraud']==0]['rule_score'].gt(0.5).mean())
```
